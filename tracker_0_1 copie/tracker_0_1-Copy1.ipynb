```python
f=open("parcours_03.csv",'r')
content=f.readlines()
f.close()
```

```python
print(content) # contenu du fichier parcours_03.csv
```

```python
fw=open('parcours_03_log.csv','w')
hr=7
mn=23
sec=0
for line in content:
    lin2=line.strip('\n)(')
    if lin2 == "" :  # le saut de ligne dans le fichier parcours_03.csv est utilisé ici pour signifier un arret de 10h
        hr = hr+10
        continue
#    content3 = content3 + '   [' +  lin2 + '],\n' 
    sec=sec+30
    if sec > 59:
        mn = mn + 1
        sec = sec - 60
    if mn > 59 :
        hr = hr + 1
        mn = mn -60
    lin2="01/02/2019"+" , "+str(hr)+":"+str(mn)+":"+str(sec)+" , "+lin2
    fw.write(lin2 + "\n")
    
fw.close()
```
